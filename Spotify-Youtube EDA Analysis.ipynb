# Top 10 Most Listened to Artists (Spotify) and Most Viewed Artists (YouTube)

```python
sns.set(style="whitegrid")

fig, ax = plt.subplots(1, 2, figsize=(18, 10))

sns.barplot(
    x='Stream',
    y='Artist',
    data=top10_listened_artist,
    color='skyblue',
    ax=ax[0],
    orient='h',
    width=0.5 
)

for p in ax[0].patches:
    width = p.get_width()
    ax[0].annotate(
        f'{width:,.0f}',
        (width + 5000, p.get_y() + p.get_height() / 2),
        ha='left',
        va='center',
        xytext=(10, 0),
        textcoords='offset points'
    )

ax[0].set_title('Top 10 Artists on Spotify')
ax[0].set_xlabel('Total Streams')
ax[0].set_ylabel('Artist')

sns.barplot(
    x='Views',
    y='Artist',
    data=top10_viewed_artist,
    color='skyblue',
    ax=ax[1],
    orient='h',
    width=0.5
)

for p in ax[1].patches:
    width = p.get_width()
    ax[1].annotate(
        f'{width:,.0f}',
        (width + 5000, p.get_y() + p.get_height() / 2),
        ha='left',
        va='center',
        xytext=(10, 0),
        textcoords='offset points'
    )

ax[1].set_title('Top 10 Artists on YouTube')
ax[1].set_xlabel('Total Views')
ax[1].set_ylabel('Artist')


plt.tight_layout()
plt.show()
```

# Structural Features of Songs

```python
music_structural_columns = ['Instrumentalness', 'Speechiness', 'Acousticness', 'Liveness']

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))  
fig.tight_layout(pad=5.0) 

for i, column in enumerate(music_structural_columns):
    ax = axes[i // 2, i % 2]
    sns.histplot(df[column], kde=True, ax=ax, color='skyblue')
    
    mean_value = df[column].mean()
    ax.axvline(mean_value, color='red', linestyle='dashed', linewidth=1)
    
    ax.set_title(f'Distribution of {column.capitalize()}')
    ax.set_xlabel(column.capitalize())
    ax.set_ylabel('Frequency')
    ax.legend([f'Average: {mean_value:.2f}'], loc='upper right')

plt.tight_layout()
plt.show()
```

**Structural Features of Songs**

- **Instrumental Ratio:** The instrumental ratio is extremely low, indicating that the vast majority of songs contain vocals. However, the table shows a long tail, suggesting that there are some instrumental songs as outliers.  

- **Speechiness:** The speech value of songs is low and exhibits a right-skewed distribution. The small proportion of speech values reflects the lyrics in the songs.  

- **Acoustic Distribution:** Acoustic values appear to be evenly distributed. While a notable portion of songs has high acoustic values, the table also shows that the proportion of live recordings is very low.  

- **Studio vs. Live Recordings:** Most songs are studio recordings, as indicated by the low percentage of live recordings.

# Relationship Between Songs' View Counts, Likes, and Comment Counts

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Views'], df['Likes'], color='skyblue', alpha=0.7, edgecolors='black')
plt.title('Relationship between Views and Likes')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.grid(True, linestyle='--', alpha=0.7)


z = np.polyfit(df['Views'], df['Likes'], 1)
p = np.poly1d(z)
plt.plot(df['Views'],p(df['Views']),"r--")

plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(df['Views'], df['Comments'], color='lightgreen', alpha=0.7, edgecolors='black')
plt.title('Relationship between Views and Comments')
plt.xlabel('Views')
plt.ylabel('Comments')
plt.grid(True, linestyle='--', alpha=0.7)


z = np.polyfit(df['Views'], df['Comments'], 1)
p = np.poly1d(z)
plt.plot(df['Views'],p(df['Views']),"r--")

plt.show()
```

**Analysis of Relationship Between View Counts, Likes, and Comments**

- **Views and Likes:** There is a relatively strong positive relationship between view counts and likes. As the number of views increases, the number of likes tends to rise as well. While there are outliers, a general positive correlation can be observed.

- **Views and Comments:** A very distinct positive relationship between views and comments is not evident. An increase in views does not necessarily lead to an increase in comments. Some songs with relatively low view counts have received a disproportionately high number of comments.

- **Comment Content:** The nature of the comments could provide further insights; however, only the comment count data is available. Considering not all comments are positive, establishing a clear relationship based solely on comment counts is not feasible.



# Average Length of Songs

```python
df['Duration_Mn'] = df['Duration_ms'] / 60000

duration_mean = df['Duration_Mn'].mean()

plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(df['Duration_Mn'], bins=30, color='blue', edgecolor='k', alpha=0.7)

plt.axvline(duration_mean, color='red', linestyle='dashed', linewidth=1)
plt.text(duration_mean + 0.5, plt.ylim()[1] * 0.9, f'Mean: {duration_mean:.2f} min', 
         color='red', fontsize=12, ha='left')

plt.title('Distribution of Track Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

# Relationship Between Song Length, Likes, and Comments

```python
df['Duration_category'] = pd.cut(df['Duration_ms'], bins=[0, 120000, 180000, 240000, 300000], 
                                 labels=['<=2min', '2-3min', '3-4min', '>=4min'])

heatmap_data = df.groupby('Duration_category')[['Likes', 'Comments']].sum().reset_index()

heatmap_data['Likes'] = heatmap_data['Likes'] / 1e9
heatmap_data['Comments'] = heatmap_data['Comments'] / 1e9

annot = heatmap_data.set_index('Duration_category').T
for col in annot.columns:
    annot[col] = annot[col].apply(lambda x: f'{x:.2f}B')

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data.set_index('Duration_category').T, annot=annot, cmap='coolwarm', linewidths=0.5, fmt="")
plt.title('Heatmap of Total Likes and Comments by Song Duration Category (in billions)')
plt.xlabel('Duration Category')
plt.ylabel('Metric')
plt.show()
```

**Average Song Length and Its Relationship with Likes and Comments**

- **Average Song Length:** The average song length is 3.75 minutes. A significant majority of songs fall within the 3-4 minute range.  

- **Likes:** Songs with a length of 3-4 minutes tend to have a higher number of likes. However, for songs shorter than 3 minutes or longer than 4 minutes, the relationship with likes appears to be weak.  

- **Comments:** Similarly, there is a weak relationship between song length and the number of comments. Songs in the 3-4 minute range seem to have slightly higher comment counts, but overall, the correlation between song length and comments is minimal.

- **Conclusion:** While the 3-4 minute length appears optimal for engagement, the general correlation between song length, likes, and comments is weak, indicating other factors play a larger role in driving audience interaction.

# Contribution of the Most Viewed 10 Songs to Their Channels

```python
top_10_viewed_songs = df.sort_values(by='Views', ascending=False).head(10)

top_10_songs_channels = top_10_viewed_songs[['Track', 'Channel', 'Views']]

top_channels = top_10_songs_channels['Channel'].unique()
channels_views = df[df['Channel'].isin(top_channels)].groupby('Channel')['Views'].sum().reset_index()

top_10_songs_views_per_channel = top_10_songs_channels.groupby('Channel')['Views'].sum().reset_index()
channels_views = channels_views.merge(top_10_songs_views_per_channel, on='Channel', suffixes=('_total', '_top10'))

channels_views['Top10_Contribution'] = channels_views['Views_top10'] / channels_views['Views_total'] * 100

channels_views = channels_views.sort_values('Top10_Contribution', ascending=False)
sorted_channels = channels_views['Channel']

fig, ax1 = plt.subplots(figsize=(14, 8))


sns.barplot(x='Views_total', y='Channel', data=channels_views, color='b', label='Total Views', ax=ax1, order=sorted_channels)
sns.barplot(x='Views_top10', y='Channel', data=channels_views, color='orange', label='Top 10 Songs Views', ax=ax1, order=sorted_channels)

for i, (total, top10) in enumerate(zip(channels_views['Views_total'], channels_views['Views_top10'])):
    ax1.text(total, i, f'{total/1e6:.2f}M', ha='left', va='center', fontsize=10, color='black', alpha=0.7)
    ax1.text(top10, i, f'{top10/1e6:.2f}M', ha='right', va='center', fontsize=10, color='black', alpha=0.7)
    
ax2 = ax1.twiny()
ax2.set_xlim(ax1.get_xlim())
sns.scatterplot(x='Top10_Contribution', y='Channel', data=channels_views, color='red', label='% Top 10 Contribution', ax=ax2)


for i, contribution in enumerate(channels_views['Top10_Contribution']):
    ax2.text(contribution + 1, i, f'{contribution:.2f}%', ha='left', va='center', fontsize=16, color='red', alpha=0.7)

ax1.set_title('Top 10 Most Viewed Songs and Their Contribution to Total Views of Channels')
ax1.set_xlabel('Views')
ax1.set_ylabel('Channel')
ax1.legend(loc='lower right')
ax2.set_xlabel('Top 10 Contribution (%)')
ax2.legend(loc='upper right')

plt.show()
```

**Contribution of Top 10 Songs to Their Channels**

- **Wiz Khalifa:** The song *See You Again* contributes a remarkable **97%** to Wiz Khalifa's channel viewership. This indicates that the success of this song is almost entirely responsible for the channel's overall performance.

- **Mark Ronson and Luis Fonsi:** These artists rank second, with their top songs contributing over **70%** to their channels' total views. This highlights the significant impact of their hit tracks on channel success.

- **Artists' Channels:** For channels like those of Wiz Khalifa, Mark Ronson, and Luis Fonsi, the channels are artist-centric. Their popular songs have not only made them famous but have also significantly boosted the success and visibility of their YouTube channels.

- **Ed Sheeran, Katy Perry, and Maroon 5:** These artists have a more balanced contribution from their top songs, with the impact on their channels being below **50%**. This is considered normal, as these artists have multiple popular songs, and their channel views are distributed across several successful tracks.

- **Conclusion:**  
  - For some artists, a single viral hit can drive the majority of a channel's success (*See You Again* for Wiz Khalifa).  
  - For others, like Ed Sheeran and Katy Perry, the success is more diversified across multiple popular songs, leading to a more balanced view distribution across their channel.  
  - These insights highlight how top songs can define or complement the success of an artist's YouTube channel.

# Distribution of Songs by Licensing and Official Status

```python
licensed_counts = df['Licensed'].value_counts(dropna=False)

licensed_counts.index = ['Licensed' if val == True else 'Unlicensed' 
                                    if val == False else 'Unknown' 
                                    for val in licensed_counts.index]


official_counts = df['official_video'].value_counts(dropna=False)

official_counts.index = ['Official' if val == True else 'Unofficial' 
                                    if val == False else 'Unknown' 
                                    for val in official_counts.index]

colors_licensed = ['#66b3ff', '#ff9999', '#c2c2f0']
colors_official = ['#66b3ff', '#ff9999', '#c2c2f0']


autopct_format = lambda p: f'{p:.1f}%'


plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
licensed_counts.plot.pie(autopct=autopct_format, colors=colors_licensed, startangle=140, textprops={'fontsize': 14})
plt.title('Distribution of Licensed Songs')
plt.ylabel('')


plt.subplot(1, 2, 2)
official_counts.plot.pie(autopct=autopct_format, colors=colors_official, startangle=140, textprops={'fontsize': 14})
plt.title('Distribution of Official Videos')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

**Distribution of Songs by Licensing and Official Status**

- **Licensed Songs:**  
  - **68%** of the songs in the dataset are licensed. This indicates that a majority of the content has the necessary permissions and rights for distribution.  

- **Official Songs:**  
  - **76%** of the songs are marked as official. These songs represent content directly uploaded by artists or associated labels, typically showcasing higher production quality and branding.  

- **Insight:**  
  - The proportion of licensed songs (**68%**) is slightly lower compared to official songs (**76%**). This suggests that not all official content is licensed, possibly due to differences in copyright requirements or independent releases. Licensed songs are assumed to be copyright-protected, which explains their lower proportion relative to official content.

- **Unlicensed or Unofficial Songs:**  
  - A smaller proportion of the dataset consists of unlicensed or unofficial content. These could include covers, remixes, or user-uploaded content that may not have formal approval.  

- **Key Observations:**
  - **Licensed and Official Content:** These dominate both views and engagement metrics due to the trust and authenticity associated with verified uploads.
  - **Unlicensed or Unofficial Content:** While less common, these can occasionally gain traction, particularly for niche or creative reinterpretations of popular songs.

- **Conclusion:**  
  - The majority of songs in the dataset are both licensed and official, underscoring the importance of authenticity and proper licensing in driving engagement and distribution success.
